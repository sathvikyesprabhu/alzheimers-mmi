```python
import os
import shutil
import numpy as np
```

```python
num = 0
mat = np.zeros((3,2))
for folder_idx,folder in enumerate(["train","test"]):
    for cls_idx,subfolder in enumerate(["AD","CN","MCI"]):
        files_MRI = os.listdir('MRI/' + folder + '/' + subfolder)
        files_EHR = os.listdir('EHR_old/' + folder + '/' + subfolder)
        
        for file1 in files_MRI:
            part2 = file1.split('_')[2]
            RID = part2.split('.')[0]
            
            if RID[0]=='0':
                RID = RID[1:]
            
            for file2 in files_EHR:
                RID2 = file2.split('_')[0]
                if RID == RID2:
                    num += 1
                    mat[cls_idx, folder_idx] += 1
                    print(folder, subfolder, file1, file2)
                    # shutil.copy('EHR_old/' + folder + '/' + subfolder + '/' + file2,
                    #            'EHR/' + folder + '/' + subfolder + '/' + file1)
```
